<div style="background: #f0f4ff; border-left: 6px solid #4361ee; padding: 20px; border-radius: 8px; margin: 10px 0;">
  <h3 style="color: #4361ee; margin: 0 0 10px;">📌 What are Control Statements?</h3>
  <p style="color: #333; margin: 0; line-height: 1.8;">Control statements are <strong>special instructions</strong> that determine the <strong>flow of execution</strong> in a Python program. Without them, code runs top-to-bottom with no decisions, no choices, and no repetition. Control statements let you:</p>
  <ul style="color: #444; margin: 10px 0 0 0; line-height: 2;">
    <li>✅ <strong>Decide</strong> — run a block of code only when a condition is met</li>
    <li>🔁 <strong>Repeat</strong> — run a block of code multiple times (loops)</li>
    <li>🔀 <strong>Select</strong> — choose one path from many options</li>
  </ul>
  <p style="color: #555; margin: 10px 0 0; font-style: italic;">In this session we focus on <strong>decision-making</strong> statements.</p>
</div>

<div style="background: linear-gradient(90deg, #f72585, #7209b7); padding: 18px 25px; border-radius: 10px; margin-top: 30px;">
  <h2 style="color: white; margin: 0; font-size: 1.6em;">🔀 Section 1: if — elif — else Statement</h2>
</div>

<div style="background: #fff8f9; border: 1px solid #f72585; padding: 20px; border-radius: 0 0 10px 10px; margin-top: -5px;">
  <h4 style="color: #7209b7; margin-top: 0;">📖 Definition</h4>
  <p style="color: #444; line-height: 1.8;">
    The <code>if-elif-else</code> statement is a <strong>decision-making structure</strong> that tests multiple conditions in order:
  </p>
  <ul style="color: #444; line-height: 2;">
    <li>🟢 <code><strong>if</strong></code> — checks the <strong>first</strong> condition. If <code>True</code>, its block runs and the rest are skipped.</li>
    <li>🟡 <code><strong>elif</strong></code> — short for <em>else if</em>. Checks the <strong>next</strong> condition only if the previous was <code>False</code>.</li>
    <li>🔴 <code><strong>else</strong></code> — runs when <strong>none</strong> of the above conditions are <code>True</code>. Acts as a safety net.</li>
  </ul>
  <div style="background: #1a1a2e; color: #64ffda; padding: 14px 18px; border-radius: 8px; font-family: monospace; font-size: 0.95em; margin-top: 10px;">
    <span style="color: #f72585;">if</span> condition_1:<br>
    &nbsp;&nbsp;&nbsp;&nbsp;<span style="color: #a8b2d8;"># runs if condition_1 is True</span><br>
    <span style="color: #f72585;">elif</span> condition_2:<br>
    &nbsp;&nbsp;&nbsp;&nbsp;<span style="color: #a8b2d8;"># runs if condition_2 is True</span><br>
    <span style="color: #f72585;">else</span>:<br>
    &nbsp;&nbsp;&nbsp;&nbsp;<span style="color: #a8b2d8;"># runs if no condition above is True</span>
  </div>
</div>

<div style="background: linear-gradient(90deg, #4361ee, #3a0ca3); padding: 18px 25px; border-radius: 10px; margin-top: 40px;">
  <h2 style="color: white; margin: 0; font-size: 1.6em;">🪆 Section 2: Nested if-else Statement</h2>
</div>

<div style="background: #f0f4ff; border: 1px solid #4361ee; padding: 20px; border-radius: 0 0 10px 10px; margin-top: -5px;">
  <h4 style="color: #3a0ca3; margin-top: 0;">📖 Definition</h4>
  <p style="color: #444; line-height: 1.8;">
    A <strong>nested if-else</strong> is an <code>if-else</code> statement placed <strong>inside another</strong> <code>if-else</code> statement.
    It is used when a second decision must only be made <em>after</em> a first condition has already been checked and passed.
  </p>
  <p style="color: #555; line-height: 1.8; margin: 0;">Think of it like a <strong>two-step security check</strong>: first you check if someone has a pass, and <em>then</em> you check if their pass is valid. The inner check only happens if the outer check passes.</p>
  <div style="background: #1a1a2e; color: #64ffda; padding: 14px 18px; border-radius: 8px; font-family: monospace; font-size: 0.95em; margin-top: 14px;">
    <span style="color: #4361ee;">if</span> outer_condition:<br>
    &nbsp;&nbsp;&nbsp;&nbsp;<span style="color: #4361ee;">if</span> inner_condition:<br>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<span style="color: #a8b2d8;"># both conditions are True</span><br>
    &nbsp;&nbsp;&nbsp;&nbsp;<span style="color: #4361ee;">else</span>:<br>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<span style="color: #a8b2d8;"># outer True, inner False</span><br>
    <span style="color: #4361ee;">else</span>:<br>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<span style="color: #a8b2d8;"># outer condition was False</span>
  </div>
</div>

<div style="background: #e8eeff; border-left: 5px solid #4361ee; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #4361ee;">Example 1.</strong> <span style="color: #333;">Check if a number is negative, zero, or positive using nested if-else.</span>
</div>

```python
# ── Example 1: Positive, Negative, or Zero ─────────────────────────
# OUTER check: Is the number >= 0? (i.e., not negative)
#   INNER check: Is it > 0 (positive) or == 0 (zero)?
# If outer check fails → the number must be negative.

num = int(input("Enter a number: "))

if num >= 0:                          # OUTER: is the number 0 or more?
    if num > 0:                       # INNER: is it strictly positive?
        print(f"{num} is Positive ✅")
    else:                             # INNER else: num must be exactly 0
        print(f"{num} is Zero ⭕")
else:                                 # OUTER else: num is less than 0
    print(f"{num} is Negative ❌")
```

<div style="background: #e8eeff; border-left: 5px solid #4361ee; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #4361ee;">Example 2.</strong> <span style="color: #333;">Check if a year is after 2000. If yes, determine whether it is an even or odd year.</span>
</div>

```python
# ── Example 2: Year Checker ────────────────────────────────────────
# OUTER check 1: Is the year a valid positive number?
#   OUTER check 2: Is it after the year 2000?
#     INNER check: Is it even (divisible by 2) or odd?

year = int(input("Enter a year: "))

if year > 0:                          # OUTER: is the year a positive number?
    if year > 2000:                   # INNER level 1: is it after 2000?
        if year % 2 == 0:             # INNER level 2: is it divisible by 2?
            print(f"{year} is after 2000 and it is an EVEN year 📅")
        else:
            print(f"{year} is after 2000 and it is an ODD year 📅")
    else:
        print(f"{year} is before or equal to 2000 ⏮️")
else:
    print(f"{year} is not a valid year ❌  (must be a positive number)")
```

<div style="background: #e8eeff; border-left: 5px solid #4361ee; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #4361ee;">Example 3.</strong> <span style="color: #333;">Check if a person has a ticket. If yes, then verify whether the seat number is valid (1–100).</span>
</div>

```python
# ── Example 3: Ticket & Seat Validator ────────────────────────────
# OUTER check: Does the person have a ticket? (yes/no)
#   INNER check: If yes, is the seat number in the valid range (1–100)?
# The inner check only runs if the person actually has a ticket.

ticket = input("Does the person have a ticket? (yes / no): ").strip().lower()

if ticket == "yes" or ticket == "no":     # OUTER: is the input valid?
    if ticket == "yes":                   # INNER: person HAS a ticket
        seat = int(input("Enter seat number (1–100): "))
        if 1 <= seat <= 100:              # DEEP INNER: is seat number in range?
            print(f"Seat {seat} — Valid seat. Welcome aboard! 🎟️")
        else:
            print(f"Seat {seat} — Invalid seat number ❌  (must be between 1 and 100)")
    else:                                  # ticket == "no"
        print("Person does not have a ticket. Entry denied. 🚫")
else:
    print("Invalid input! Please type 'yes' or 'no' only.")
```

<div style="background: #e8eeff; border-left: 5px solid #4361ee; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #4361ee;">Example 4.</strong> <span style="color: #333;">Check if a number is divisible by 3. If yes, also check whether it is divisible by 9.</span>
</div>

```python
# ── Example 4: Divisibility Check (3 and 9) ───────────────────────
# KEY INSIGHT: Every number divisible by 9 is ALSO divisible by 3.
#              But NOT every number divisible by 3 is divisible by 9.
#              So we check divisibility by 3 FIRST (outer),
#              then check divisibility by 9 inside (inner).

number = int(input("Enter a positive number: "))

if number > 0:                         # OUTER: only work with positive numbers
    if number % 3 == 0:                # INNER: is it divisible by 3?
        print(f"{number} IS divisible by 3 ✅")
        if number % 9 == 0:            # DEEP INNER: also divisible by 9?
            print(f"{number} is ALSO divisible by 9 ✅✅")
        else:
            print(f"{number} is NOT divisible by 9 ❌")
    else:
        print(f"{number} is NOT divisible by 3 ❌")
elif number == 0:
    print("Zero cannot be tested for divisibility.")
else:
    print(f"{number} is a negative number. Please enter a positive number.")
```

<div style="background: #e8eeff; border-left: 5px solid #4361ee; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #4361ee;">Example 5.</strong> <span style="color: #333;">Check if temperature is above 20°C. If yes, determine whether it is also above 35°C (hot) or between 20–35°C (warm).</span>
</div>

```python
# ── Example 5: Temperature Range Checker ─────────────────────────
# OUTER check: Is the temperature above 20°C?
#   INNER check: Is it below 35°C (warm) or 35°C and above (hot)?
# If the temperature is 20°C or below → it is cold / cool.

temperature = float(input("Enter temperature in degrees Celsius (°C): "))

if temperature > 20:                    # OUTER: is it above 20°C?
    if temperature < 35:                # INNER: is it below 35°C?
        print(f"{temperature}°C — Warm 🌤️  (between 20°C and 35°C)")
    else:                               # INNER else: 35°C or above
        print(f"{temperature}°C — Hot 🔥  (35°C or above)")
else:                                   # OUTER else: 20°C or below
    print(f"{temperature}°C — Cold/Cool ❄️  (20°C or below)")
```

<div style="background: linear-gradient(90deg, #06d6a0, #0a9396); padding: 18px 25px; border-radius: 10px; margin-top: 40px;">
  <h2 style="color: white; margin: 0; font-size: 1.6em;">🎯 Section 3: Match-Case Statement</h2>
</div>

<div style="background: #f0fff9; border: 1px solid #06d6a0; padding: 20px; border-radius: 0 0 10px 10px; margin-top: -5px;">
  <h4 style="color: #0a9396; margin-top: 0;">📖 Definition</h4>
  <p style="color: #444; line-height: 1.8;">
    The <code>match</code> statement, introduced in <strong>Python 3.10</strong>, compares a single variable against multiple <strong>fixed cases</strong> and executes the matching block. It is Python's version of a <em>switch-case</em> found in other languages like C, Java, or JavaScript.
  </p>

  <h4 style="color: #0a9396;">🔑 Key Points</h4>
  <ul style="color: #444; line-height: 2; margin: 0;">
    <li>✅ Available from <strong>Python 3.10</strong> onwards</li>
    <li>🎯 Best used when comparing <strong>one variable</strong> against many <strong>exact values</strong></li>
    <li>🛡️ <code>case _:</code> is the <strong>default / wildcard</strong> — runs when no other case matches</li>
    <li>📖 Much <strong>cleaner and more readable</strong> than a long chain of <code>if-elif</code> for fixed values</li>
    <li>⚡ Python checks each case in order and <strong>stops</strong> at the first match</li>
  </ul>

  <h4 style="color: #0a9396; margin-top: 16px;">📐 Syntax</h4>
  <div style="background: #1a1a2e; color: #64ffda; padding: 14px 18px; border-radius: 8px; font-family: monospace; font-size: 0.95em;">
    <span style="color: #06d6a0;">match</span> variable:<br>
    &nbsp;&nbsp;&nbsp;&nbsp;<span style="color: #06d6a0;">case</span> value1:<br>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<span style="color: #a8b2d8;"># runs if variable == value1</span><br>
    &nbsp;&nbsp;&nbsp;&nbsp;<span style="color: #06d6a0;">case</span> value2:<br>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<span style="color: #a8b2d8;"># runs if variable == value2</span><br>
    &nbsp;&nbsp;&nbsp;&nbsp;<span style="color: #06d6a0;">case</span> _: <span style="color: #a8b2d8;"># default — runs if no case matched</span><br>
    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<span style="color: #a8b2d8;"># fallback code here</span>
  </div>
</div>

<div style="background: #d8f8f0; border-left: 5px solid #06d6a0; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #0a9396;">Example 1.</strong> <span style="color: #333;">Enter a day number (1–7) and print the corresponding day name. Print an error for any other number.</span>
</div>

```python
# ── Match Example 1: Day Number to Day Name ────────────────────────
# We match the integer 'day' against exact values 1 through 7.
# 'case _:' is the DEFAULT — it catches anything not matched above.

day = int(input("Enter a day number (1–7): "))

match day:
    case 1:
        print("📅 Day 1 → Monday")
    case 2:
        print("📅 Day 2 → Tuesday")
    case 3:
        print("📅 Day 3 → Wednesday")
    case 4:
        print("📅 Day 4 → Thursday")
    case 5:
        print("📅 Day 5 → Friday")
    case 6:
        print("📅 Day 6 → Saturday")
    case 7:
        print("📅 Day 7 → Sunday")
    case _:                            # DEFAULT: none of the above matched
        print(f"❌ '{day}' is an invalid day number. Please enter 1–7.")
```

<div style="background: #d8f8f0; border-left: 5px solid #06d6a0; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #0a9396;">Example 2.</strong> <span style="color: #333;">Build a simple calculator using match-case. Take two numbers and an operator (+, -, *, /), then perform the operation.</span>
</div>

```python
# ── Match Example 2: Simple Calculator ────────────────────────────
# We match the 'operator' string against the four math symbols.
# Inside the division case, we add an extra check to prevent
# dividing by zero, which would cause a crash.

num1     = float(input("Enter first  number : "))
num2     = float(input("Enter second number : "))
operator = input("Enter operator (+, -, *, /): ")

match operator:
    case "+":                          # Addition
        result = num1 + num2
        print(f"✅ {num1} + {num2} = {result}")

    case "-":                          # Subtraction
        result = num1 - num2
        print(f"✅ {num1} - {num2} = {result}")

    case "*":                          # Multiplication
        result = num1 * num2
        print(f"✅ {num1} × {num2} = {result}")

    case "/":                          # Division (with zero-check)
        if num2 != 0:
            result = num1 / num2
            print(f"✅ {num1} ÷ {num2} = {result}")
        else:
            print("❌ Error: Cannot divide by zero!")

    case _:                            # DEFAULT: unrecognized operator
        print(f"❌ '{operator}' is not a valid operator. Use +, -, *, or /")
```

<div style="background: #d8f8f0; border-left: 5px solid #06d6a0; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #0a9396;">Example 3.</strong> <span style="color: #333;">Check a student's grade (A, B, C, D) using match-case and print the corresponding performance message.</span>
</div>

```python
# ── Match Example 3: Student Grade Checker ─────────────────────────
# We match the grade letter directly.
# '.upper()' converts lowercase input (a, b, c, d) to uppercase
# so the user can type either 'a' or 'A' and it still works.
# 'case _:' handles any grade letter that is not A, B, C, or D.

grade = input("Enter your grade (A / B / C / D): ").strip().upper()

match grade:
    case "A":                          # Grade A → highest achievement
        print("🌟 Grade A — Excellent! Outstanding performance!")
    case "B":                          # Grade B → above average
        print("👍 Grade B — Good! Keep it up!")
    case "C":                          # Grade C → average
        print("📘 Grade C — Average. There is room to improve.")
    case "D":                          # Grade D → below average / fail
        print("⚠️  Grade D — Fail. Please work harder next time.")
    case _:                            # DEFAULT: any other input
        print(f"❌ '{grade}' is not a valid grade. Please enter A, B, C, or D.")
```

  <h4 style="color: #e85d04; margin-top: 0;">📖 When to use which?</h4>
  <p style="color: #444; line-height: 1.8;">
    Both structures make decisions, but they serve different purposes.
    Use <strong>if-elif-else</strong> for flexible, complex conditions involving ranges or expressions.
    Use <strong>match-case</strong> for clean, readable comparisons against a fixed set of exact values.
  </p>
