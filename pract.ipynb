## -> WHILE LOOP:
#### wHILE LOOP is used to repeatdely execute a given block of code  as long as given condition is true.

```python
# this example simply demostrate the while loop structure and how its work
count = 0
while count<=4:
    print(count)
    count +=  1
```

```python
name ="Maroof Ahmed lone"
count =3
while True:
    if name[0]=='M':
        name=name[2:]
    elif name[-1]=='b':
        name = name[:2]
    else:
        count+=1
        break
    print(name)
    print(count)
    
```

## topic: functions

A function is a bblock of code which only run when it called, u can pass data ,known as parammeter,into the function. A function can return data as a result

```python
# a function that only greet display ton user
def greet(name):
    print("HELLO,Welcome ",name)
    
greet("asad")

# function that  add two value passing by parameter

def add(a,b):
    sum = a+b
    print(f"sum of {a} and {b} is : {sum}")
add(5,7)

# function that multipy ,div, and mod the values passing by parameter
def func(a,b):
    mult = a*b
    mod = a%b
    div = a/b
    div2 = a//b
    
    print(f"the multiplkication of the {a} and {b} is: {mult}",
           f" the mod of {a} and {b} is : {mod}",
          f"div of {a} and {b} is : {div}",
        f"divsion withouut decimal value is  {div2}")
    
    
func(18,19)
```
