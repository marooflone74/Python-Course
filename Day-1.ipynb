  <h2 style="color: white; margin: 0; font-size: 1.6em;">📦 Section 1: Variables in Python</h2>
</div>

<div style="background: #fff8f9; border: 1px solid #f72585; padding: 20px; border-radius: 0 0 10px 10px; margin-top: -5px;">
  <h4 style="color: #7209b7; margin-top: 0;">❓ Q1. What is a variable in Python? Write the syntax and give one example.</h4>
  <h4 style="color: #f72585; margin-bottom: 6px;">📖 Answer</h4>
  <p style="color: #444; line-height: 1.8; margin: 0;">
    A <strong>variable</strong> is a <em>named storage location</em> in computer memory that holds a value. Think of it like a <strong>labelled box</strong> — you give the box a name, and you can put any value inside it. The value can be changed (updated) at any time during the program.
  </p>
  <div style="background: #1a1a2e; color: #64ffda; padding: 14px 18px; border-radius: 8px; font-family: monospace; font-size: 0.95em; margin-top: 14px;">
    <span style="color: #a8b2d8;"># Syntax:</span><br>
    <span style="color: #f72585;">variable_name</span> <span style="color: white;">= value</span>
  </div>
</div>

```python
# ── Q1: Creating a Variable ────────────────────────────────────────
# We create a variable named 'salary_emp1' and assign it the value 230000.
# The '=' sign is the ASSIGNMENT operator — it stores the value on the
# right side into the variable on the left side.

salary_emp1 = 230000     # Stores the integer 230000 into 'salary_emp1'

print(f"Employee Salary : {salary_emp1}")
```

<div style="background: #fff8f9; border: 1px solid #f72585; padding: 20px; border-radius: 10px; margin-top: 20px;">
  <h4 style="color: #7209b7; margin-top: 0;">❓ Q2. Write any three rules for naming a variable in Python. Give one correct and one incorrect example.</h4>
  <h4 style="color: #f72585; margin-bottom: 6px;">📖 Answer</h4>
  <p style="color: #444; line-height: 1.8;"><strong>Rules for naming variables in Python:</strong></p>
  <ol style="color: #444; line-height: 2.2; margin: 0; padding-left: 20px;">
    <li>A variable name must <strong>start with a letter</strong> (a–z, A–Z) or an <strong>underscore (_)</strong>. It <em>cannot</em> start with a number.</li>
    <li>A variable name can only contain <strong>letters, digits, and underscores</strong>. Special symbols like <code>@, #, $, !</code> are <strong>not allowed</strong>.</li>
    <li>Variable names are <strong>case-sensitive</strong> — <code>age</code>, <code>Age</code>, and <code>AGE</code> are three <em>different</em> variables.</li>
    <li>A variable name <strong>cannot be a Python keyword</strong> (reserved word) like <code>if</code>, <code>for</code>, <code>while</code>, <code>print</code>, etc.</li>
  </ol>
</div>

```python
# ── Q2A: VALID variable name examples ─────────────────────────────
# ✅ Starts with a letter, uses underscore, no special symbols

roll_no     = 42          # ✅ lowercase with underscore
StudentName = "Ali"       # ✅ starts with uppercase letter
_score      = 98.5        # ✅ starts with underscore
total2024   = 1500        # ✅ number at the END is allowed

print(f"Roll No      : {roll_no}")
print(f"Student Name : {StudentName}")
print(f"Score        : {_score}")
print(f"Total 2024   : {total2024}")
```

```python
# ── Q2B: INVALID variable name examples ───────────────────────────
# ❌ These names break Python naming rules.
#    They are shown as COMMENTS because running them causes a SyntaxError.

# ❌ 123rollno  = 42      → starts with a digit (NOT allowed)
# ❌ roll-no    = 42      → '-' is a minus operator, not underscore
# ❌ roll@no    = 42      → '@' is a special symbol (NOT allowed)
# ❌ for        = 42      → 'for' is a reserved Python keyword

print("Invalid names shown above as comments to avoid SyntaxError.")
print("Remember: Start with a letter or underscore, no special symbols!")
```

<div style="background: #e8eeff; border-left: 5px solid #4361ee; padding: 14px 18px; border-radius: 6px; margin-top: 22px;">
  <h4 style="color: #3a0ca3; margin: 0 0 8px;">❓ Q3. What is the integer data type? Write one Python example.</h4>
  <p style="color: #444; line-height: 1.8; margin: 0;">
    The <code><strong>int</strong></code> (integer) data type stores <strong>whole numbers</strong> — positive, negative, or zero — without any decimal point. Use it when your value will always be a complete number (e.g., age, roll number, quantity).
  </p>
  <p style="color: #555; margin: 8px 0 0; font-style: italic;">Examples of integers: <code>0, 5, -10, 1000, 999999</code></p>
</div>

```python
# ── Q3: Integer (int) Data Type ────────────────────────────────────
# An integer holds a whole number — no decimal point.
# Python automatically treats a plain number like 23000 as int.

employee_salary = 23000      # Whole number → int type
student_age     = 20         # Age is always a whole number
roll_number     = 7          # Roll no. is a whole number

print(f"Employee Salary : {employee_salary}  → type: {type(employee_salary)}")
print(f"Student Age     : {student_age}       → type: {type(student_age)}")
print(f"Roll Number     : {roll_number}        → type: {type(roll_number)}")
```

<div style="background: #e8eeff; border-left: 5px solid #4361ee; padding: 14px 18px; border-radius: 6px; margin-top: 20px;">
  <h4 style="color: #3a0ca3; margin: 0 0 8px;">❓ Q4. What is the float data type? Write one Python example using a decimal value.</h4>
  <p style="color: #444; line-height: 1.8; margin: 0;">
    The <code><strong>float</strong></code> data type stores <strong>numbers with a decimal point</strong> (fractional values). Use it for measurements, prices, percentages, or any value that needs precision beyond whole numbers.
  </p>
  <p style="color: #555; margin: 8px 0 0; font-style: italic;">Examples of floats: <code>3.14, 99.99, -0.5, 44.96, 2.0</code></p>
</div>

```python
# ── Q4: Float Data Type ────────────────────────────────────────────
# A float holds a number with a decimal point.
# Python automatically treats a number like 44.96 as float.

temperature     = 44.96      # Temperature with decimal → float
product_price   = 199.99     # Price with decimal → float
pi_value        = 3.14159    # Mathematical constant → float

print(f"Temperature   : {temperature}   → type: {type(temperature)}")
print(f"Product Price : {product_price}  → type: {type(product_price)}")
print(f"Pi Value      : {pi_value}  → type: {type(pi_value)}")
```

<div style="background: #e8eeff; border-left: 5px solid #4361ee; padding: 14px 18px; border-radius: 6px; margin-top: 20px;">
  <h4 style="color: #3a0ca3; margin: 0 0 8px;">❓ Q5. What is the string data type? Write two ways to create a string in Python with examples.</h4>
  <p style="color: #444; line-height: 1.8; margin: 0;">
    A <code><strong>str</strong></code> (string) is a <strong>sequence of characters</strong> enclosed inside quotes. It is used to store text — names, messages, addresses, or any readable content. In Python, a string can be created using:
  </p>
  <ol style="color: #444; line-height: 2.2; margin: 8px 0 0; padding-left: 20px;">
    <li><strong>Single quotes</strong> — <code>'Hello'</code></li>
    <li><strong>Double quotes</strong> — <code>"Hello"</code></li>
    <li><strong>Triple quotes</strong> — <code>'''Hello'''</code> or <code>"""Hello"""</code> (for multi-line strings)</li>
  </ol>
</div>

```python
# ── Q5: String (str) Data Type ─────────────────────────────────────
# Strings store text — any characters between quotes.
# Both single and double quotes work the same way.

# Method 1: Using single quotes
student_name  = 'Ali'

# Method 2: Using double quotes
course_name   = "Programming Fundamentals"

# Method 3: Using triple quotes (allows multi-line text)
description   = """Python is a powerful,
beginner-friendly language."""

print(f"Student Name : {student_name}    → type: {type(student_name)}")
print(f"Course Name  : {course_name} → type: {type(course_name)}")
print(f"Description  :\n{description}")
```

<div style="background: #e8eeff; border-left: 5px solid #4361ee; padding: 14px 18px; border-radius: 6px; margin-top: 20px;">
  <h4 style="color: #3a0ca3; margin: 0 0 8px;">❓ Q6. What is the boolean data type? Write its possible values and one example.</h4>
  <p style="color: #444; line-height: 1.8; margin: 0;">
    The <code><strong>bool</strong></code> (boolean) data type can hold only <strong>two possible values</strong>:
    <code>True</code> or <code>False</code>. It is mainly used in <strong>conditions, loops, and flags</strong>
    to represent a yes/no or on/off state. Note: <code>True</code> and <code>False</code> must be capitalized — Python is case-sensitive.
  </p>
  <table style="border-collapse: collapse; margin-top: 10px; font-size: 0.9em;">
    <tr style="background: #3a0ca3; color: white;">
      <th style="padding: 7px 14px;">Value</th>
      <th style="padding: 7px 14px;">Meaning</th>
      <th style="padding: 7px 14px;">Numeric Equivalent</th>
    </tr>
    <tr style="background: #323439;"><td style="padding: 7px 14px;"><code>True</code></td><td>Yes / On / Correct</td><td><code>1</code></td></tr>
    <tr style="background: #26272a;"><td style="padding: 7px 14px;"><code>False</code></td><td>No / Off / Incorrect</td><td><code>0</code></td></tr>
  </table>
</div>

```python
# ── Q6: Boolean (bool) Data Type ──────────────────────────────────
# A boolean stores only True or False.
# It is used as a FLAG — to track a yes/no state in your program.

is_student     = True        # This person IS a student
is_graduated   = False       # This person has NOT graduated yet
is_logged_in   = True        # User IS currently logged in

print(f"Is Student   : {is_student}    → type: {type(is_student)}")
print(f"Is Graduated : {is_graduated}   → type: {type(is_graduated)}")
print(f"Is Logged In : {is_logged_in}    → type: {type(is_logged_in)}")
```

  <h2 style="color: white; margin: 0; font-size: 1.6em;">🛠️ Section 3: Built-in Functions — type(), input(), Type Casting</h2>

<div style="background: #f0fff9; border: 1px solid #06d6a0; padding: 14px 18px; border-radius: 6px; margin-top: 10px;">
  <h4 style="color: #0a9396; margin: 0 0 8px;">❓ Q7. What does the type() function do in Python? Explain with an example.</h4>
  <p style="color: #444; line-height: 1.8; margin: 0;">
    The <code><strong>type()</strong></code> function is a built-in Python function that <strong>returns the data type</strong> of any variable or value. It is very useful for <strong>debugging</strong> — when you are unsure what type a variable is, you can use <code>type()</code> to check it.
  </p>
  <div style="background: #1a1a2e; color: #64ffda; padding: 12px 16px; border-radius: 6px; font-family: monospace; margin-top: 10px;">
    <span style="color: #a8b2d8;"># Syntax:</span><br>
    <span style="color: #06d6a0;">type</span>( variable_or_value )
  </div>
  <p style="color: #555; margin: 8px 0 0; font-style: italic;">Output format: <code>&lt;class 'int'&gt;</code>, <code>&lt;class 'str'&gt;</code>, <code>&lt;class 'float'&gt;</code>, etc.</p>
</div>

```python
# ── Q7: The type() Function ────────────────────────────────────────
# type() tells you the data type of any variable.
# Wrap it inside print() to see the output.

engine_no    = 634100          # int
temperature  = 36.6            # float
student_name = "Mujtaba"       # str
is_passed    = True            # bool

print(f"engine_no    = {engine_no}     → {type(engine_no)}")
print(f"temperature  = {temperature}        → {type(temperature)}")
print(f"student_name = {student_name}      → {type(student_name)}")
print(f"is_passed    = {is_passed}       → {type(is_passed)}")
```

<div style="background: #f0fff9; border: 1px solid #06d6a0; padding: 14px 18px; border-radius: 6px; margin-top: 20px;">
  <h4 style="color: #0a9396; margin: 0 0 8px;">❓ Q8. What is the input() function in Python? What is the default data type of input()? Give an example.</h4>
  <p style="color: #444; line-height: 1.8; margin: 0;">
    The <code><strong>input()</strong></code> function <strong>pauses the program</strong> and waits for the user to type something on the keyboard. Whatever the user types is stored as the return value. The <strong>default data type</strong> of <code>input()</code> is always <code>str</code> (string) — even if the user types a number like <code>25</code>, it is stored as <code>'25'</code> (a string).
  </p>
  <p style="color: #c0392b; background: #fff0f0; padding: 8px 12px; border-radius: 6px; margin: 10px 0 0; font-size: 0.9em;">
    ⚠️ <strong>Important:</strong> If you need the input as a number, you must convert it using <code>int()</code> or <code>float()</code> — this is called <strong>Type Casting</strong> (covered in Q9).
  </p>
</div>

```python
# ── Q8: The input() Function ───────────────────────────────────────
# input() takes a prompt string as an argument (what to display to user),
# waits for the user to type, and returns whatever they typed as a STRING.

student_name = input("Enter your name: ")     # Always returns str

print(f"Name entered  : {student_name}")
print(f"Data type     : {type(student_name)}")    # Always <class 'str'>
```

<div style="background: #f0fff9; border: 1px solid #06d6a0; padding: 14px 18px; border-radius: 6px; margin-top: 20px;">
  <h4 style="color: #0a9396; margin: 0 0 8px;">❓ Q9. What is type casting? Write a Python example to convert a string into an integer.</h4>
  <p style="color: #444; line-height: 1.8; margin: 0;">
    <strong>Type casting</strong> (also called <em>type conversion</em>) means <strong>explicitly converting</strong> a value from one data type to another. Python provides built-in functions for this:
  </p>
  <table style="border-collapse: collapse; margin-top: 10px; font-size: 0.9em; width: 100%;">
    <tr style="background: #0a9396; color: white;">
      <th style="padding: 8px 14px; text-align: left;">Function</th>
      <th style="padding: 8px 14px; text-align: left;">Converts To</th>
      <th style="padding: 8px 14px; text-align: left;">Example</th>
    </tr>
    <tr style="background: #d8f8f0;"><td><code>int()</code></td><td>Integer</td><td><code>int("25") → 25</code></td></tr>
    <tr style="background: #b8f0e4;"><td><code>float()</code></td><td>Float (decimal)</td><td><code>float("3.14") → 3.14</code></td></tr>
    <tr style="background: #d8f8f0;"><td><code>str()</code></td><td>String (text)</td><td><code>str(100) → '100'</code></td></tr>
    <tr style="background: #b8f0e4;"><td><code>bool()</code></td><td>Boolean</td><td><code>bool(0) → False</code></td></tr>
  </table>
</div>

```python
# ── Q9: Type Casting ───────────────────────────────────────────────
# By default, input() always returns a STRING.
# To do math, we must convert it to int or float using type casting.

# ── Without type casting ──
name = input("Enter student name   : ")     # stays as str ✅ (correct)
print(f"Name       : {name}  → type: {type(name)}")

# ── With type casting to int ──
# int() wraps around input() and converts the string to an integer
roll_no = int(input("Enter student roll no: "))   # '42' → 42 (int)
print(f"Roll No    : {roll_no}   → type: {type(roll_no)}")

# ── With type casting to float ──
marks = float(input("Enter student marks  : "))   # '95.5' → 95.5 (float)
print(f"Marks      : {marks}  → type: {type(marks)}")
```

<div style="background: #f0fff9; border: 1px solid #06d6a0; padding: 14px 18px; border-radius: 6px; margin-top: 20px;">
  <h4 style="color: #0a9396; margin: 0 0 8px;">❓ Q10. Write a Python program that takes a number from the user and prints its data type.</h4>
</div>

```python
# ── Q10: Take a Number from User and Print its Data Type ──────────
# We take a salary as a float (to allow decimal values like 5000.50),
# then print both the value and its data type using type().

# Step 1: Take input and cast to float
salary = float(input("Enter employee salary: "))    # str → float

# Step 2: Print the value
print(f"\nSalary entered : {salary}")

# Step 3: Print the data type
print(f"Data type      : {type(salary)}")

# Bonus: Show what type it would be WITHOUT casting
raw_input = str(salary)    # convert back to string to demonstrate
print(f"\n(Without casting, input() returns: {type(raw_input)})")
```
