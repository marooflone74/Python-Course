<div style="border-left: 4px solid #a8edea; background: #f0fdfc; border-radius: 0 12px 12px 0; padding: 22px 28px; margin: 8px 0;">
  <div style="color:#a8edea; font-size:11px; font-weight:700; letter-spacing:2px; text-transform:uppercase; margin-bottom:10px;">📖 Definition</div>
  <p style="color:#1a1a2e; font-size:15px; line-height:1.8; margin:0;">
    A <strong style='color:#0f3460;'>for loop</strong> is used to <strong>iterate over a sequence</strong> — such as a list, tuple, string, or range. It executes a block of code <strong>once for each item</strong> in the sequence, automatically moving to the next item after each iteration.
  </p>
</div>

<div style="background:#1a1a2e; border-radius: 12px; padding: 22px 28px; margin: 10px 0;">
  <div style="color:#a8edea; font-size:11px; font-weight:700; letter-spacing:2px; text-transform:uppercase; margin-bottom:14px;">📝 Syntax</div>

    <span style='color:#6c7086;'># block of code to execute</span></pre>
</div>

<div style="background:#f8f9ff; border:1px solid #e0e4f5; border-radius: 12px; padding: 18px 24px; margin: 10px 0;">

<div style="display:flex; align-items:center; gap:12px; margin: 20px 0 4px 0;">
  <div style="background:#a8edea; color:#0f3460; font-size:11px; font-weight:800; letter-spacing:1px; padding:4px 12px; border-radius:20px;">EXAMPLE 01</div>
  <span style="color:#444; font-size:14px; font-weight:600;">Print numbers 1 to 5 using <code>range()</code></span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 1 | for Loop — Print numbers 1 to 5
# ─────────────────────────────────────────────────────────

# range(1, 6) generates: 1, 2, 3, 4, 5
for number in range(1, 6):
    print(number)   # prints each number on a new line
```

<div style="display:flex; align-items:center; gap:12px; margin: 20px 0 4px 0;">
  <div style="background:#a8edea; color:#0f3460; font-size:11px; font-weight:800; letter-spacing:1px; padding:4px 12px; border-radius:20px;">EXAMPLE 02</div>
  <span style="color:#444; font-size:14px; font-weight:600;">Iterate over a list of fruits</span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 2 | for Loop — Iterate over a list
# ─────────────────────────────────────────────────────────

fruits = ["Apple", "Banana", "Cherry", "Mango", "Grapes"]

for fruit in fruits:           # loop through each item in the list
    print("Fruit:", fruit)     # print each fruit name
```

<div style="display:flex; align-items:center; gap:12px; margin: 20px 0 4px 0;">
  <div style="background:#a8edea; color:#0f3460; font-size:11px; font-weight:800; letter-spacing:1px; padding:4px 12px; border-radius:20px;">EXAMPLE 03</div>
  <span style="color:#444; font-size:14px; font-weight:600;">Iterate over characters in a string</span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 3 | for Loop — Loop through a string
# ─────────────────────────────────────────────────────────

name = "Python"

# A string is a sequence of characters — loop visits each one
for letter in name:
    print(letter)    # prints one character per line
```

<div style="display:flex; align-items:center; gap:12px; margin: 20px 0 4px 0;">
  <div style="background:#a8edea; color:#0f3460; font-size:11px; font-weight:800; letter-spacing:1px; padding:4px 12px; border-radius:20px;">EXAMPLE 04</div>
  <span style="color:#444; font-size:14px; font-weight:600;">Calculate the sum of numbers in a list</span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 4 | for Loop — Sum of all numbers in a list
# ─────────────────────────────────────────────────────────

numbers = [10, 20, 30, 40, 50]
total   = 0    # start with zero

for num in numbers:
    total = total + num    # add each number to running total

print("Sum of all numbers:", total)    # Output: 150
```

<div style="display:flex; align-items:center; gap:12px; margin: 20px 0 4px 0;">
  <div style="background:#a8edea; color:#0f3460; font-size:11px; font-weight:800; letter-spacing:1px; padding:4px 12px; border-radius:20px;">EXAMPLE 05</div>
  <span style="color:#444; font-size:14px; font-weight:600;">Multiplication table using for loop</span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 5 | for Loop — Multiplication Table
# ─────────────────────────────────────────────────────────

table_of = 5    # generate table for number 5

print(f"── Multiplication Table of {table_of} ──")
for i in range(1, 11):                         # i goes from 1 to 10
    result = table_of * i                       # calculate product
    print(f"  {table_of} x {i:2d} = {result}") # print formatted row
```

<div style="border-left: 4px solid #fbc2eb; background: #fdf0f8; border-radius: 0 12px 12px 0; padding: 22px 28px; margin: 8px 0;">
  <div style="color:#c2185b; font-size:11px; font-weight:700; letter-spacing:2px; text-transform:uppercase; margin-bottom:10px;">📖 Definition</div>
  <p style="color:#1a1a2e; font-size:15px; line-height:1.8; margin:0;">
    A <strong style='color:#3a0062;'>while loop</strong> keeps executing a block of code <strong>as long as a condition remains <code>True</code></strong>. It is ideal when the number of iterations is <strong>not known in advance</strong>. The condition is checked <em>before</em> each iteration.
  </p>
</div>

<div style="background:#1a1a2e; border-radius: 12px; padding: 22px 28px; margin: 10px 0;">
  <div style="color:#fbc2eb; font-size:11px; font-weight:700; letter-spacing:2px; text-transform:uppercase; margin-bottom:14px;">📝 Syntax</div>
  <pre style="margin:0; color:#cdd6f4; font-size:14px; font-family: 'Courier New', monospace; line-height:1.7;"><span style='color:#89b4fa;'>while</span> <span style='color:#f9ca24;'>condition</span>:
    <span style='color:#6c7086;'># block of code to execute</span>
    <span style='color:#6c7086;'># ⚠️ always update the variable to avoid infinite loop!</span></pre>
</div>

<div style="background:#fff8e1; border:1px solid #ffe082; border-radius: 12px; padding: 14px 20px; margin: 10px 0; display:flex; gap:12px; align-items:flex-start;">
  <span style="font-size:20px;">⚠️</span>
  <p style="margin:0; color:#5d4037; font-size:14px; line-height:1.7;"><strong>Important:</strong> Always update the loop variable inside a <code>while</code> loop. If the condition never becomes <code>False</code>, the loop will run forever (infinite loop), causing your program to hang.</p>
</div>

<div style="background:#f8f9ff; border:1px solid #e0e4f5; border-radius: 12px; padding: 18px 24px; margin: 10px 0;">
  <table style="width:100%; border-collapse:collapse; font-size:14px;">

      <th style="padding:10px 14px; text-align:left; color:#3a0062; font-weight:700;">Term</th>

    <tr style="border-bottom:1px solid #e0e4f5;">
      <td style="padding:10px 14px;"><code style='background:#fdf0f8; color:#3a0062; padding:2px 8px; border-radius:4px;'>condition</code></td>
      <td style="padding:10px 14px; color:#333;">An expression that evaluates to True or False</td>
    </tr>
    <tr style="border-bottom:1px solid #e0e4f5;">
      <td style="padding:10px 14px;"><code style='background:#fdf0f8; color:#3a0062; padding:2px 8px; border-radius:4px;">True</code></td>
      <td style="padding:10px 14px; color:#333;">Loop body executes — continues looping</td>
    </tr>
    <tr>
      <td style="padding:10px 14px;"><code style='background:#fdf0f8; color:#3a0062; padding:2px 8px; border-radius:4px;">False</code></td>
      <td style="padding:10px 14px; color:#333;">Loop stops — program moves on</td>
    </tr>
  </table>

<div style="display:flex; align-items:center; gap:12px; margin: 20px 0 4px 0;">
  <div style="background:#fbc2eb; color:#3a0062; font-size:11px; font-weight:800; letter-spacing:1px; padding:4px 12px; border-radius:20px;">EXAMPLE 01</div>
  <span style="color:#444; font-size:14px; font-weight:600;">Print numbers 1 to 5 using while loop</span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 1 | while Loop — Print 1 to 5
# ─────────────────────────────────────────────────────────

count = 1    # initialize counter variable

while count <= 5:          # keep looping while count is <= 5
    print(count)           # print current value
    count = count + 1      # increment — prevents infinite loop!
```

<div style="display:flex; align-items:center; gap:12px; margin: 20px 0 4px 0;">
  <div style="background:#fbc2eb; color:#3a0062; font-size:11px; font-weight:800; letter-spacing:1px; padding:4px 12px; border-radius:20px;">EXAMPLE 02</div>
  <span style="color:#444; font-size:14px; font-weight:600;">Countdown timer</span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 2 | while Loop — Countdown Timer
# ─────────────────────────────────────────────────────────

countdown = 5    # start counting from 5

while countdown > 0:              # keep looping while positive
    print(f"  🕐 Countdown: {countdown}")
    countdown -= 1                # decrease by 1 (same as countdown = countdown - 1)

print("🚀 Launch!")               # prints after the loop ends
```

<div style="display:flex; align-items:center; gap:12px; margin: 20px 0 4px 0;">
  <div style="background:#fbc2eb; color:#3a0062; font-size:11px; font-weight:800; letter-spacing:1px; padding:4px 12px; border-radius:20px;">EXAMPLE 03</div>
  <span style="color:#444; font-size:14px; font-weight:600;">Sum of digits of a number</span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 3 | while Loop — Sum of Digits (1234 → 1+2+3+4 = 10)
# ─────────────────────────────────────────────────────────

number    = 1234    # the number whose digits we want to add
digit_sum = 0       # running total

while number > 0:
    digit      = number % 10       # extract last digit  (1234 % 10 = 4)
    digit_sum += digit             # add it to sum
    number     = number // 10      # remove last digit   (1234 // 10 = 123)

print("Sum of digits:", digit_sum)    # Output: 10
print("last num:", number)
```

<div style="display:flex; align-items:center; gap:12px; margin: 20px 0 4px 0;">
  <div style="background:#fbc2eb; color:#3a0062; font-size:11px; font-weight:800; letter-spacing:1px; padding:4px 12px; border-radius:20px;">EXAMPLE 04</div>
  <span style="color:#444; font-size:14px; font-weight:600;">Print even numbers from 2 to 10</span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 4 | while Loop — Even Numbers from 2 to 10
# ─────────────────────────────────────────────────────────

num = 2    # start at first even number

print("Even numbers from 2 to 10:")
while num <= 10:    # loop until we pass 10
    print(num)      # print current even number
    num += 2        # jump by 2 to skip odd numbers
```

<div style="display:flex; align-items:center; gap:12px; margin: 20px 0 4px 0;">
  <div style="background:#fbc2eb; color:#3a0062; font-size:11px; font-weight:800; letter-spacing:1px; padding:4px 12px; border-radius:20px;">EXAMPLE 05</div>
  <span style="color:#444; font-size:14px; font-weight:600;">Find first power of 2 greater than 100</span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 5 | while Loop — First Power of 2 > 100
# ─────────────────────────────────────────────────────────

power = 1    # start at 2^0 = 1

# keep doubling until we exceed 100
while power <= 100:
    power = power * 2    # double the value each time

print("First power of 2 greater than 100:", power)    # Output: 128
```

  <span style="color:#444; font-size:14px; font-weight:600;">Stop the loop when number 4 is reached</span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 1 | break — Stop loop at number 4
# ─────────────────────────────────────────────────────────

for num in range(1, 10):       # loop from 1 to 9
    if num == 4:               # when we reach 4 ...
        print("Found 4! Stopping the loop.")
        break                  # exit loop — 5,6,7,8,9 will NOT print
    print(num)                 # prints 1, 2, 3

print("Loop ended.")           # this runs AFTER the break
```

<div style="display:flex; align-items:center; gap:12px; margin: 20px 0 4px 0;">
  <div style="background:#f9ca24; color:#3d1a00; font-size:11px; font-weight:800; letter-spacing:1px; padding:4px 12px; border-radius:20px;">EXAMPLE 02</div>
  <span style="color:#444; font-size:14px; font-weight:600;">Search for a name in a list</span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 2 | break — Name Search in a List
# ─────────────────────────────────────────────────────────

students = ["Ali", "Sara", "John", "Zara", "Mike"]
search   = "John"    # the name we are looking for

for student in students:
    if student == search:         # if found ...
        print(f"✅ '{search}' found in the list!")
        break                     # no need to keep searching
    print(f"Checking: {student}") # still searching ...
```

<div style="display:flex; align-items:center; gap:12px; margin: 20px 0 4px 0;">
  <div style="background:#f9ca24; color:#3d1a00; font-size:11px; font-weight:800; letter-spacing:1px; padding:4px 12px; border-radius:20px;">EXAMPLE 03</div>
  <span style="color:#444; font-size:14px; font-weight:600;">Stop when total exceeds 50</span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 3 | break — Stop Accumulating When Total > 50
# ─────────────────────────────────────────────────────────

total = 0
num   = 1

while True:                  # infinite loop — only break stops it
    total += num             # keep adding numbers: 1, 2, 3 ...
    if total > 50:           # once total exceeds 50 ...
        print(f"Total reached {total} at number {num}. Stopping!")
        break                # exit immediately
    num += 1
```

<div style="display:flex; align-items:center; gap:12px; margin: 20px 0 4px 0;">
  <div style="background:#f9ca24; color:#3d1a00; font-size:11px; font-weight:800; letter-spacing:1px; padding:4px 12px; border-radius:20px;">EXAMPLE 04</div>
  <span style="color:#444; font-size:14px; font-weight:600;">Find the first even number in a list</span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 4 | break — Find First Even Number
# ─────────────────────────────────────────────────────────

numbers = [3, 7, 11, 8, 15, 20]

for num in numbers:
    if num % 2 == 0:                       # check if number is even
        print(f"First even number: {num}") # found it!
        break                              # stop — we only need the first one
    print(f"{num} is odd, searching...")
```

<div style="display:flex; align-items:center; gap:12px; margin: 20px 0 4px 0;">
  <div style="background:#f9ca24; color:#3d1a00; font-size:11px; font-weight:800; letter-spacing:1px; padding:4px 12px; border-radius:20px;">EXAMPLE 05</div>
  <span style="color:#444; font-size:14px; font-weight:600;">ATM PIN system with maximum 3 attempts</span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 5 | break — ATM PIN with Max 3 Tries
# ─────────────────────────────────────────────────────────

correct_pin = 1234    # the correct PIN
attempts    = 0       # track number of attempts

while attempts < 3:                          # maximum 3 tries allowed
    pin = int(input("Enter your PIN: "))     # ask user for PIN
    attempts += 1

    if pin == correct_pin:                   # correct PIN entered
        print("✅ Access Granted! Welcome.")
        break                               # exit loop — no more attempts needed
    else:
        print(f"❌ Wrong PIN! Attempts left: {3 - attempts}")
else:
    # 'else' after while runs ONLY if loop finished WITHOUT a break
    print("🔒 Account Locked after 3 failed attempts.")
```

<div style="background: linear-gradient(135deg, #003d1a, #1a1a2e); border-radius: 16px; padding: 28px 36px; margin: 40px 0 0 0; display:flex; align-items:center; gap:18px;">

    <h2 style="color:#ffffff; font-size:28px; font-weight:800; margin:0;">The <code style='background:rgba(107,203,119,0.15); color:#6bcb77; padding:2px 10px; border-radius:6px; font-size:26px;'>continue</code> Statement</h2>
  </div>
</div>

<div style="border-left: 4px solid #6bcb77; background: #f0fdf4; border-radius: 0 12px 12px 0; padding: 22px 28px; margin: 8px 0;">
  <div style="color:#2e7d32; font-size:11px; font-weight:700; letter-spacing:2px; text-transform:uppercase; margin-bottom:10px;">📖 Definition</div>
  <p style="color:#1a1a2e; font-size:15px; line-height:1.8; margin:0;">
    The <strong style='color:#003d1a;'>continue</strong> statement <strong>skips the remainder</strong> of the current iteration and immediately moves to the <strong>next iteration</strong> of the loop. Unlike <code>break</code>, it does <em>not</em> stop the loop — the loop continues normally.
  </p>
</div>

<div style="background:#1a1a2e; border-radius: 12px; padding: 22px 28px; margin: 10px 0;">
  <div style="color:#6bcb77; font-size:11px; font-weight:700; letter-spacing:2px; text-transform:uppercase; margin-bottom:14px;">📝 Syntax</div>
  <pre style="margin:0; color:#cdd6f4; font-size:14px; font-family: 'Courier New', monospace; line-height:1.7;"><span style='color:#89b4fa;'>for</span> <span style='color:#a6e3a1;'>variable</span> <span style='color:#89b4fa;'>in</span> <span style='color:#f9ca24;'>sequence</span>:
    <span style='color:#89b4fa;'>if</span> <span style='color:#f38ba8;'>condition</span>:
        <span style='color:#6bcb77;'>continue</span>    <span style='color:#6c7086;'># skip current iteration, go to next</span>
    <span style='color:#6c7086;'># this code is SKIPPED when continue triggers</span></pre>
</div>

<div style="background:#f8f9ff; border:1px solid #e0e4f5; border-radius: 12px; padding: 18px 24px; margin: 10px 0;">
  <table style="width:100%; border-collapse:collapse; font-size:14px;">
    <tr style="background:#f0fdf4;">
      <th style="padding:10px 14px; text-align:left; color:#003d1a; font-weight:700;">Behaviour</th>
      <th style="padding:10px 14px; text-align:left; color:#003d1a; font-weight:700;">Detail</th>
    </tr>
    <tr style="border-bottom:1px solid #e0e4f5;">
      <td style="padding:10px 14px; font-weight:600; color:#2e7d32;">Skips iteration</td>
      <td style="padding:10px 14px; color:#333;">Code below <code>continue</code> is skipped for this step</td>
    </tr>
    <tr style="border-bottom:1px solid #e0e4f5;">
      <td style="padding:10px 14px; font-weight:600; color:#2e7d32;">Loop continues</td>
      <td style="padding:10px 14px; color:#333;">Loop moves on to the next item normally</td>
    </tr>
    <tr>
      <td style="padding:10px 14px; font-weight:600; color:#2e7d32;">Works in</td>
      <td style="padding:10px 14px; color:#333;">Both <code>for</code> and <code>while</code> loops</td>
    </tr>
  </table>

<div style="display:flex; align-items:center; gap:12px; margin: 20px 0 4px 0;">
  <div style="background:#6bcb77; color:#003d1a; font-size:11px; font-weight:800; letter-spacing:1px; padding:4px 12px; border-radius:20px;">EXAMPLE 01</div>
  <span style="color:#444; font-size:14px; font-weight:600;">Skip number 3 while printing 1 to 6</span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 1 | continue — Skip number 3
# ─────────────────────────────────────────────────────────

for num in range(1, 7):    # loop from 1 to 6
    if num == 3:           # when num is 3 ...
        continue           # skip this iteration (3 won't print)
    print(num)             # prints: 1, 2, 4, 5, 6
```

<div style="display:flex; align-items:center; gap:12px; margin: 20px 0 4px 0;">
  <div style="background:#6bcb77; color:#003d1a; font-size:11px; font-weight:800; letter-spacing:1px; padding:4px 12px; border-radius:20px;">EXAMPLE 02</div>
  <span style="color:#444; font-size:14px; font-weight:600;">Print only even numbers using continue</span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 2 | continue — Print Only Even Numbers
# ─────────────────────────────────────────────────────────

print("Even numbers from 1 to 10:")
for num in range(1, 11):
    if num % 2 != 0:     # if number is ODD ...
        continue         # skip it — jump to next iteration
    print(num)           # only even numbers reach this line
```

<div style="display:flex; align-items:center; gap:12px; margin: 20px 0 4px 0;">
  <div style="background:#6bcb77; color:#003d1a; font-size:11px; font-weight:800; letter-spacing:1px; padding:4px 12px; border-radius:20px;">EXAMPLE 03</div>
  <span style="color:#444; font-size:14px; font-weight:600;">Skip negative numbers when summing</span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 3 | continue — Sum Only Positive Numbers
# ─────────────────────────────────────────────────────────

numbers      = [10, -5, 30, -8, 20, -3, 15]
positive_sum = 0    # holds sum of positive numbers only

for num in numbers:
    if num < 0:                      # if the number is negative ...
        continue                     # skip it — don't add to sum
    positive_sum += num              # only positives are added

print("Sum of positive numbers:", positive_sum)    # Output: 75
```

<div style="display:flex; align-items:center; gap:12px; margin: 20px 0 4px 0;">
  <div style="background:#6bcb77; color:#003d1a; font-size:11px; font-weight:800; letter-spacing:1px; padding:4px 12px; border-radius:20px;">EXAMPLE 04</div>
  <span style="color:#444; font-size:14px; font-weight:600;">Skip vowels and print only consonants</span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 4 | continue — Print Only Consonants
# ─────────────────────────────────────────────────────────

word   = "programming"
vowels = "aeiou"           # all vowel characters

print("Consonants in 'programming':")
for letter in word:
    if letter in vowels:   # if letter is a vowel ...
        continue           # skip it
    print(letter)          # only consonants get printed
```

<div style="display:flex; align-items:center; gap:12px; margin: 20px 0 4px 0;">
  <div style="background:#6bcb77; color:#003d1a; font-size:11px; font-weight:800; letter-spacing:1px; padding:4px 12px; border-radius:20px;">EXAMPLE 05</div>
  <span style="color:#444; font-size:14px; font-weight:600;">Skip empty strings from a list</span>
</div>

```python
# ─────────────────────────────────────────────────────────
# Example 5 | continue — Filter Out Empty Strings
# ─────────────────────────────────────────────────────────

names = ["Alice", "", "Bob", "", "Charlie", "", "David"]

print("Valid names:")
for name in names:
    if name == "":           # if the name is empty ...
        continue             # skip empty strings
    print(f"  ✅ {name}")    # only valid names reach here
```
