<div style="background: #fff3cd; border-left: 5px solid #f72585; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #f72585;">Q1.</strong> <span style="color: #333;">Write a Python program to add two numbers and print the result.</span>
</div>

```python
# ── Q1: Addition of Two Numbers ──────────────────────────────────
# We use the '+' arithmetic operator to add two numbers together.

num1 = 5         # First number
num2 = 6         # Second number

result = num1 + num2   # '+' operator adds num1 and num2

print(f"Number 1  : {num1}")
print(f"Number 2  : {num2}")
print(f"Sum       : {num1} + {num2} = {result}")
```

<div style="background: #fff3cd; border-left: 5px solid #f72585; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #f72585;">Q2.</strong> <span style="color: #333;">Write a program to find the remainder when one number is divided by another.</span>
</div>

```python
# ── Q2: Modulus Operator (%) ──────────────────────────────────────
# The '%' (modulus) operator returns the REMAINDER after division.
# Example: 7 ÷ 4 = 1 remainder 3  → so 7 % 4 = 3

dividend = 7     # The number to be divided
divisor  = 4     # The number we divide by

remainder = dividend % divisor   # '%' gives the leftover after division

print(f"Dividend  : {dividend}")
print(f"Divisor   : {divisor}")
print(f"Remainder : {dividend} % {divisor} = {remainder}")
```

<div style="background: #fff3cd; border-left: 5px solid #f72585; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #f72585;">Q3.</strong> <span style="color: #333;">Write a program to calculate the square of a number using the power operator.</span>
</div>

```python
# ── Q3: Power / Exponentiation Operator (**) ─────────────────────
# The '**' operator raises a number to the power of another.
# To find the SQUARE, we raise a number to the power of 2.
# Example: 5² = 5 ** 2 = 25

number = 5       # The base number
power  = 2       # Squaring means power = 2

square = number ** power    # '**' operator: number raised to 'power'

print(f"Number    : {number}")
print(f"Power     : {power}")
print(f"Square    : {number} ** {power} = {square}")
```

<div style="background: #e8eeff; border-left: 5px solid #4361ee; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #4361ee;">Q4.</strong> <span style="color: #333;">Write a program that assigns a value to a variable and then increases it by 5 using an assignment operator.</span>
</div>

```python
# ── Q4: += (Add and Assign Operator) ─────────────────────────────
# '+=' adds a value to the variable and saves the result back.
# It is a shortcut: instead of writing  score = score + 5
#                   we write            score += 5

score = 10       # Initial value of the variable

print(f"Score before '+=' operator : {score}")

score += 5       # Same as: score = score + 5

print(f"Score after  '+= 5'         : {score}")
```

<div style="background: #e8eeff; border-left: 5px solid #4361ee; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #4361ee;">Q5.</strong> <span style="color: #333;">Write a program that multiplies a variable by 2 using an assignment operator.</span>
</div>

```python
# ── Q5: *= (Multiply and Assign Operator) ─────────────────────────
# '*=' multiplies the variable by a value and stores the result.
# Shortcut: instead of  price = price * 2
#           we write    price *= 2

price = 15       # Initial price

print(f"Price before '*=' operator : {price}")

price *= 2       # Same as: price = price * 2  (doubles the price)

print(f"Price after  '*= 2'         : {price}")
```

<div style="background: #e8eeff; border-left: 5px solid #4361ee; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #4361ee;">Q6.</strong> <span style="color: #333;">Write a program that subtracts 10 from a variable using an assignment operator.</span>
</div>

```python
# ── Q6: -= (Subtract and Assign Operator) ─────────────────────────
# '-=' subtracts a value from the variable and saves the result.
# Shortcut: instead of  budget = budget - 10
#           we write    budget -= 10

budget = 100     # Initial budget amount

print(f"Budget before '-=' operator : {budget}")

budget -= 10     # Same as: budget = budget - 10  (spent 10)

print(f"Budget after  '-= 10'        : {budget}")
```

<div style="background: #d8f8f0; border-left: 5px solid #06d6a0; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #0a9396;">Q7.</strong> <span style="color: #333;">Write a program to check whether two numbers are equal or not.</span>
</div>

```python
# ── Q7: == and != Comparison Operators ────────────────────────────
# '==' checks if two values are EQUAL      → returns True or False
# '!=' checks if two values are NOT EQUAL  → returns True or False

a = 5
b = 8

# Check equality
is_equal     = (a == b)   # Are a and b the same value?
is_not_equal = (a != b)   # Are a and b different values?

print(f"a = {a},  b = {b}")
print(f"a == b  (Are they equal?)     → {is_equal}")
print(f"a != b  (Are they not equal?) → {is_not_equal}")
```

<div style="background: #d8f8f0; border-left: 5px solid #06d6a0; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #0a9396;">Q8.</strong> <span style="color: #333;">Write a program to check if one number is greater than another.</span>
</div>

```python
# ── Q8: > and < Comparison Operators ─────────────────────────────
# '>'  checks if the LEFT value is GREATER than the RIGHT
# '<'  checks if the LEFT value is LESS    than the RIGHT

x = 15
y = 9

greater = (x > y)   # Is x greater than y?
lesser  = (x < y)   # Is x less than y?

print(f"x = {x},  y = {y}")
print(f"x > y  (Is x greater than y?) → {greater}")
print(f"x < y  (Is x less than y?)    → {lesser}")
```

<div style="background: #d8f8f0; border-left: 5px solid #06d6a0; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #0a9396;">Q9.</strong> <span style="color: #333;">Write a program to compare two numbers and print whether the first number is less than or equal to the second number.</span>
</div>

```python
# ── Q9: <= and >= Operators ────────────────────────────────────────
# '<=' checks if the left value is LESS THAN or EQUAL to the right
# '>=' checks if the left value is GREATER THAN or EQUAL to the right

p = 7
q = 10

less_or_equal    = (p <= q)   # Is p less than or equal to q?
greater_or_equal = (p >= q)   # Is p greater than or equal to q?

print(f"p = {p},  q = {q}")
print(f"p <= q  (Is p ≤ q?) → {less_or_equal}")
print(f"p >= q  (Is p ≥ q?) → {greater_or_equal}")
```

<div style="background: #ffe8cc; border-left: 5px solid #fb8500; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #e85d04;">Q10.</strong> <span style="color: #333;">Write a program that checks if a number is greater than 5 AND less than 20.</span>
</div>

```python
# ── Q10: 'and' Logical Operator ────────────────────────────────────
# 'and' returns True ONLY when BOTH conditions are True.
# Think of it as: "This AND that must both be true."

num = 12   # Test number — change this to try different values!

# Condition 1: num must be greater than 5
# Condition 2: num must be less than 20
# BOTH must be True for the overall result to be True

result = (num > 5) and (num < 20)

print(f"Number         : {num}")
print(f"num > 5        : {num > 5}")
print(f"num < 20       : {num < 20}")
print(f"Both true (and): {result}   ← Is {num} between 5 and 20?")
```

<div style="background: #ffe8cc; border-left: 5px solid #fb8500; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #e85d04;">Q11.</strong> <span style="color: #333;">Write a program that checks if a number is less than 10 OR greater than 50.</span>
</div>

```python
# ── Q11: 'or' Logical Operator ─────────────────────────────────────
# 'or' returns True when AT LEAST ONE condition is True.
# Think of it as: "Either this OR that (or both) is fine."

val1 = 10    # Exactly 10 — neither < 10 nor > 50
val2 = 60    # Greater than 50 — satisfies one condition

result1 = (val1 < 10) or (val1 > 50)
result2 = (val2 < 10) or (val2 > 50)

print(f"--- Test 1: val = {val1} ---")
print(f"val < 10       : {val1 < 10}")
print(f"val > 50       : {val1 > 50}")
print(f"Either true (or): {result1}\n")

print(f"--- Test 2: val = {val2} ---")
print(f"val < 10       : {val2 < 10}")
print(f"val > 50       : {val2 > 50}")
print(f"Either true (or): {result2}")
```

<div style="background: #ffe8cc; border-left: 5px solid #fb8500; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #e85d04;">Q12.</strong> <span style="color: #333;">Write a program that uses the <code>not</code> operator to reverse a condition.</span>
</div>

```python
# ── Q12: 'not' Logical Operator ────────────────────────────────────
# 'not' REVERSES or FLIPS a boolean result.
# If a condition is True  → 'not' makes it False
# If a condition is False → 'not' makes it True

n = 10

condition        = (n > 8 and n < 13)   # Is n between 8 and 13?
reversed_result  = not condition          # Flip/reverse the result

print(f"Number                      : {n}")
print(f"n > 8 and n < 13            : {condition}")
print(f"not(n > 8 and n < 13)       : {reversed_result}   ← Reversed!")
```

<div style="background: #e8deff; border-left: 5px solid #8338ec; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #8338ec;">Q13.</strong> <span style="color: #333;">Write a program to check if a number exists in a given list.</span>
</div>

```python
# ── Q13: 'in' and 'not in' with a List ────────────────────────────
# 'in'     → True if the value IS found in the list
# 'not in' → True if the value is NOT found in the list
# NOTE: Always match the data type! Integers in a list of integers.

student_marks = [60, 45, 68, 21, 34, 56, 29, 7, 36, 39]   # List of integer marks

search1 = 56    # This number IS in the list
search2 = 90    # This number is NOT in the list

print(f"Student Marks : {student_marks}\n")

print(f"Is {search1} in the list?      → {search1 in student_marks}")
print(f"Is {search1} NOT in the list?  → {search1 not in student_marks}\n")

print(f"Is {search2} in the list?      → {search2 in student_marks}")
print(f"Is {search2} NOT in the list?  → {search2 not in student_marks}")
```

<div style="background: #e8deff; border-left: 5px solid #8338ec; padding: 12px 18px; border-radius: 6px; margin-top: 20px;">
  <strong style="color: #8338ec;">Q14.</strong> <span style="color: #333;">Write a program to check if a word exists in a given collection of strings.</span>
</div>

```python
# ── Q14: Membership Check in a Set of Strings ─────────────────────
# A set is an unordered collection of unique elements.
# Membership operators work perfectly on sets too.

word_collection = {"hello", "extra", "snow", "cell"}   # A set of words

word1 = "hello"    # This word IS in the set
word2 = "world"    # This word is NOT in the set

print(f"Word Collection : {word_collection}\n")

print(f"Is '{word1}' in the set?      → {word1 in word_collection}")
print(f"Is '{word1}' NOT in the set?  → {word1 not in word_collection}\n")

print(f"Is '{word2}' in the set?      → {word2 in word_collection}")
print(f"Is '{word2}' NOT in the set?  → {word2 not in word_collection}")
```

  <strong style="color: #8338ec;">Q15.</strong> <span style="color: #333;">Write a program to check whether a value is not present in a tuple.</span>

```python
# ── Q15: 'not in' Operator with a Tuple ───────────────────────────
# A tuple is an IMMUTABLE ordered collection (cannot be changed).
# 'not in' checks if a value is ABSENT from the tuple.

numbers = (10, 20, 30, 40, 50)   # An immutable tuple of numbers

val1 = 20    # This value IS present in the tuple
val2 = 99    # This value is NOT present in the tuple

print(f"Tuple           : {numbers}\n")

print(f"Is {val1} in the tuple?      → {val1 in numbers}")
print(f"Is {val1} NOT in the tuple?  → {val1 not in numbers}\n")

print(f"Is {val2} in the tuple?      → {val2 in numbers}")
print(f"Is {val2} NOT in the tuple?  → {val2 not in numbers}")
```
